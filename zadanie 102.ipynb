```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
dane100 = np.random.normal(loc=3, scale=1, size=100)
plt.figure(figsize=(8, 5))
plt.hist(dane100, bins=20, color='skyblue', density=True)
plt.xlabel('x')
plt.ylabel('y')
plt.grid('True')
plt.show()
```

```python
dane10000 = np.random.normal(loc=3, scale=1, size=10000)
plt.figure(figsize=(8, 5))
plt.hist(dane10000, bins=20, color='skyblue', density=True)
plt.xlabel('x')
plt.ylabel('y')
plt.grid('True')
plt.show()
```

```python
mean100=np.mean(dane100)
mean100
```

```python
mean10000=np.mean(dane10000)
mean10000
```

```python
median100=np.median(dane100)
median10000=np.median(dane10000)
```

```python
mode100=stats.mode(dane100)
mode10000=stats.mode(dane10000)
mode10000
```

```python
hq0=np.quantile(dane100, 0.0)
hq1=np.quantile(dane100, 0.25)
hq3=np.quantile(dane100, 0.75)
hq4=np.quantile(dane100, 1.0)
```

```python
tq0=np.quantile(dane10000, 0.0)
tq1=np.quantile(dane10000, 0.25)
tq3=np.quantile(dane10000, 0.75)
tq4=np.quantile(dane10000, 1.0)
```

```python
iqr100=hq3-hq1
iqr10000=tq3-tq1
```

```python
plt.hist(dane100, bins=10)
plt.axvline(x = mean100, color='red', label = 'rednia')
plt.axvline(x = median100, color='green', label = 'Mediana')
plt.axvline(x = mode100[0], color='orange', label = 'Moda')
plt.axvline(x = hq0, color='black', label = 'Min')
plt.axvline(x = hq1, color='black', label = 'Q1')
plt.axvline(x = hq3, color='black', label = 'Q3')
plt.axvline(x = hq4, color='black', label = 'Max')
plt.legend(loc=(1.02,0.55))
plt.show()
```

```python
print(f"""Statystyka opisowa - rozkad normalny przy 100 losowych danych:
rednia = {mean100}
Mediana = {median100}
Warto minimalna = {hq1}
Warto maksymalna = {hq4}
Rozstp midzykwartylowy: {iqr100}
""")
```

```python
plt.hist(dane10000, bins=10)
plt.axvline(x = mean10000, color='red', label = 'rednia')
plt.axvline(x = median10000, color='green', label = 'Mediana')
plt.axvline(x = mode10000[0], color='orange', label = 'Moda')
plt.axvline(x = tq0, color='black', label = 'Min')
plt.axvline(x = tq1, color='black', label = 'Q1')
plt.axvline(x = tq3, color='black', label = 'Q3')
plt.axvline(x = tq4, color='black', label = 'Max')
plt.legend(loc=(1.02,0.55))
plt.show()
```

```python
print(f"""Statystyka opisowa - rozkad normalny przy 10000 losowych danych:
rednia = {mean10000}
Mediana = {median10000}
Warto minimalna = {tq1}
Warto maksymalna = {tq4}
Rozstp midzykwartylowy: {iqr10000}""")
```

```python
danedf = [['rednia', mean100, mean10000],['Mediana', median100,median10000], ['Warto minimalna',hq1,tq1],['Warto maksymalna',hq4,tq4],['Rozstp midzykwartylowy', iqr100, iqr10000]]
df = pd.DataFrame(danedf, columns=['Wartoci', '100', '10000'])
df
```
