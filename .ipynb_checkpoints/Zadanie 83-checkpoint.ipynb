```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

```python
df1=df.pivot_table(values='id', index=['race', 'signs_of_mental_illness'], aggfunc=np.size).rename(columns={'id':'number_of_victims'})
df1
```

```python
total=df1.groupby(level=0)['number_of_victims'].sum()
total
```

```python
ill=df1.xs(True, level='signs_of_mental_illness')['number_of_victims']
ill
```

```python
illpercent=(ill/total)*100
illpercent
```

```python
df1['Percentage']=df1.apply(lambda row: illpercent[row.name[0]], axis=1)
df1
```

```python
maxill = illpercent.max()
maxillrace=illpercent.idxmax()
print(f'Race with highest percentage of signs of mental illness is {maxillrace}')
```

```python
df['date'] = pd.to_datetime(df['date'])
df['day']=df['date'].dt.day_name()
df.head()
```

```python
dfdays=df.groupby('day').size().reset_index(name='count')
dfdays
```

```python
order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dfdays['day']=pd.Categorical(dfdays['day'], categories=order, ordered=True)
dfdays = dfdays.sort_values('day')
dfdays
```

```python
plt.bar(dfdays['day'], dfdays['count'])
```

```python
table1=pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations', header=0)
table2=pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population', header=0)
data1=table1[1]
data1.head()
```

```python
data2=table2[0]
data2.head()
```

```python
data1[['Name', 'USPS']]
data1.drop([0,1], inplace=True)
data1.rename(columns={'Name':'State'}, inplace=True)
```

```python
data2[['State','Census population, April 1, 2020 [1][2]']]
data2.rename(columns={'Census population, April 1, 2020 [1][2]':'Population'}, inplace=True)
```

```python
newdf=pd.merge(data1,data2,how='left',on='State')
newdf=newdf.rename(columns={'State':'state_name', 'USPS':'state_code'})
newdf
```

```python
df2=df.pivot_table(values='id', index=['state'], aggfunc=np.size).rename(columns={'id':'victims_per_state'})
df2=df2.reset_index()
df2.columns=['state_code','victims_per_state']
df2
```

```python
finaldf=pd.merge(newdf,df2,how='right',on='state_code')
finaldf
```

```python
finaldf['victims_per_1000']=finaldf['victims_per_state']/finaldf['Population']*1000
finaldf
```
