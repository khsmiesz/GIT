```python
def commision_fee(x):
    if x<=300:
        return 0
    elif x<=900:
        return x*0.03
    else:
        return x*0.06
```

```python
df['commision_fee']=df['Sprzedaż'].apply(lambda x: commision_fee(x)).round(2)
```

```python
df.head()
```

```python
df['Produkt_len']=df['Produkt'].apply(len)
```

```python
df.head()
```

```python
import math
```

```python
df['#_opakowań']=df['Sztuki'].apply(lambda x: math.ceil(x/5))
df.head()
```

```python
def bonus(row):
    margin=(row['Sprzedaż']-row['Koszty'])/row['Sprzedaż']
    if margin>0.55:
        return 200
    else:
        return 0
```

```python
df['Bonus']=df.apply(lambda row: bonus(row), axis=1)
df.head()
```
